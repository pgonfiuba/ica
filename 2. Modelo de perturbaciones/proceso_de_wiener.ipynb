```python
# Simulacin de un proceso de Wiener (movimiento browniano)

import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 1. Parmetros de simulacin
# -----------------------------
T = 5.0       # tiempo total
dt = 0.01     # paso de tiempo
N = int(T / dt)
t = np.linspace(0, T, N)

# -----------------------------
# 2. Simular proceso de Wiener
# -----------------------------
def simular_wiener(N, dt, trayectorias=5):
    dW = np.sqrt(dt) * np.random.randn(trayectorias, N)
    W = np.cumsum(dW, axis=1)
    return W

# Generar trayectorias
W = simular_wiener(N, dt, trayectorias=5)

# -----------------------------
# 3. Graficar trayectorias
# -----------------------------
plt.figure(figsize=(10, 5))
for i in range(W.shape[0]):
    plt.plot(t, W[i], label=f'Trayectoria {i+1}')

plt.title('Proceso de Wiener (movimiento browniano)')
plt.xlabel('Tiempo t')
plt.ylabel('W(t)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
